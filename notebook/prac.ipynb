```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
filepath = os.path.join('..','data','Internship_Selection_Dataset.csv')
df = pd.read_csv(filepath)

df.head(8)
```

```python
col = [cols for cols in df.columns if df[cols].dtype == 'O']
print(col)
```

```python
df['extracurricular'] = np.where(df['extracurricular'].str.contains("Yes"),1,0)
```

```python
col = [cols for cols in df.columns if df[cols].dtype == 'O']
print(col)
```

#### In DF['College_tier'] Encoding
{Tier 1 : 1 (3044 count), Tier 2 : 2 (4944 count), Tier 3 : 3 (2012 count)}

```python
df['college_tier'].value_counts()
```

```python
df['college_tier'] = df['college_tier'].replace("Tier 1",1)
df['college_tier'] = df['college_tier'].replace("Tier 2",2)
df['college_tier'] = df['college_tier'].replace("Tier 3",3)
```

#### In DF['placement_training'] encoding
{
    "Yes" : 1, (6876 count)
    "No" : 0, (3124 count)
}

```python
df['placement_training'].value_counts()
```

```python
df['placement_training'] = np.where(df['placement_training'].str.contains("Yes"),1,0)
```

```python
df.columns
```

```python
X = df.drop(columns='selected',axis=1)
y = df['selected']

print(X.columns)
print("\n",y.name)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=42)

X_train.shape , X_test.shape, y_train.shape, y_test.shape
```

```python
X_train.columns[:5]
```

```python
X.corr()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Fit on training data and transform both
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 1. Initialize the models
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
lr_model = LogisticRegression(max_iter=1000)

# 2. Train the models
rf_model.fit(X_train, y_train)
lr_model.fit(X_train_scaled, y_train)

# 3. Make predictions
rf_pred = rf_model.predict(X_test)
lr_pred = lr_model.predict(X_test_scaled)

# 4. Compare Accuracy
print(f"Random Forest Accuracy: {accuracy_score(y_test, rf_pred):.4f}")
print(f"Logistic Regression Accuracy: {accuracy_score(y_test, lr_pred):.4f}")
```

```python
from xgboost import XGBClassifier

# Initialize and train
xgb_model = XGBClassifier(n_estimators=100, learning_rate=0.1, max_depth=5, random_state=42)
xgb_model.fit(X_train, y_train)

print(f"XGBoost Accuracy: {xgb_model.score(X_test, y_test):.4f}")
```
